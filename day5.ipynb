```
from itertools import *
```

```
def pairwise(iterable):
    "s -> (s0,s1), (s1,s2), (s2, s3), ..."
    a, b = tee(iterable)
    next(b, None)
    return zip(a, b)
```

```
def seat_id(col, row):
  return row * 8 + col
```

```
m = 0
for line in input.split('\n'):
  p1, p2 = line[:-3], line[-3:]
  p1 = p1.replace('B', '1').replace('F', '0')
  p2 = p2.replace('R', '1').replace('L', '0')
  row = int(p1, 2)
  col = int(p2, 2)
  if seat_id(col, row) > m:
    m = seat_id(col, row)
m
```

```
def to_b(line):
  p1, p2 = line[:-3], line[-3:]
  p1 = p1.replace('B', '1').replace('F', '0')
  p2 = p2.replace('R', '1').replace('L', '0')
  row = int(p1, 2)
  col = int(p2, 2)
  return seat_id(col, row)
```

```
bins = [to_b(line) for line in input.split('\n')]
bins.sort()
```

```
for x, y in pairwise(bins):
  if y - x == 2:
    print(x, x+1, y)
```
