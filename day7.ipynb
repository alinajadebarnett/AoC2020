```
import numpy as np
import networkx as nx
import re
```

```
regex_bag = re.compile(r'(?P<char>[a-zA-Z]+ [a-zA-Z]+) ba.*')

regex_contain = re.compile(r'(?P<min>[0-9]+) (?P<char>[a-zA-Z]+ [a-zA-Z]+) bag.*')
```

```
def get_rules(input):
  rules = dict()

  for line in input.split('\n'):
    the_bag, can_contain = line.split(' contain ')
    #print(the_bag, can_contain)
    mo = regex_bag.fullmatch(the_bag)
    rules[the_bag[:-len(' bags')]] = []
    if "no other bags" in can_contain:
      rules[the_bag[:-len(' bags')]] = []
    else:
      for bag_type in can_contain.split(', '):
        mo = regex_contain.fullmatch(bag_type)
        assert mo is not None
        num_bag, type_bag = mo.group(1,2)
        num_bag = int(num_bag)
        rules[the_bag[:-len(' bags')]].append([num_bag, type_bag])
  return rules

rules = get_rules(input)
rules
```

```
D = nx.DiGraph()

for outer, inners in rules.items():
  for inner in inners:
    D.add_edge(inner[1], outer)

len(list(nx.dfs_postorder_nodes(D, source='shiny gold')))-1
```
