```
from itertools import *
from more_itertools import *
import numpy as np
import networkx as nx
import re
from sympy.ntheory import factorint
from collections import defaultdict
```

```
regex = re.compile(r'mem\[(?P<loc>[0-9]+)\] = (?P<val>[0-9]+)')

def bool_to_char(b):
  if b:
    return '1'
  else:
    return '0'

def combine(mask, value):
  mask1 = int(''.join(bool_to_char(c == 'X') for c in mask), 2)
  mask2 = int(''.join(bool_to_char(c == '1') for c in mask), 2)
  return (value & mask1) | mask2

def main(input):
  memory_bank = defaultdict(lambda: 0)
  mask = None
  for block in input.split('\nmask = '):
    line_iterator = iter(block.split('\n'))
    mask = next(line_iterator)
    for line in line_iterator:
      mo = regex.fullmatch(line)
      assert mo is not None, line
      loc, val = mo.group('loc', 'val')
      loc, val = int(loc), int(val)
      memory_bank[loc] = combine(mask, val)
  
  return sum(memory_bank.values())
```

```
regex = re.compile(r'mem\[(?P<loc>[0-9]+)\] = (?P<val>[0-9]+)')

def bool_to_char(b):
  if b:
    return '1'
  else:
    return '0'

def gen_combine(mask, address):
  
  #mask1 = int(''.join(bool_to_char(c == 'X') for c in mask), 2) # 1 at Xs
  mask2 = int(''.join(bool_to_char(c == '1') for c in mask), 2) # 1 at 1s
  address |= mask2

  idx = mask.find('X')
  # print(idx, mask, len(mask), address)
  if idx == -1:
    yield address
  else:
    address_a = address | 1 << 35 - idx
    # print('add a', address_a)
    mask = mask[:idx] + '0' + mask[idx+1:]
    yield from gen_combine(mask, address_a)
    address_b = address & ~(1 << 35 - idx)
    yield from gen_combine(mask, address_b)

def main2(input, debug=False):
  memory_bank = defaultdict(lambda: 0)
  mask = None
  for line in input.split('\n'):
    if line[1] == 'e':
      mo = regex.fullmatch(line)
      assert mo is not None, line
      loc, val = mo.group('loc', 'val')
      loc, val = int(loc), int(val)
      for addr in gen_combine(mask, loc):
        memory_bank[addr] = val
    else: # == 'a'
      mask = line[len('mask = '):]
      if debug:
        print('mask is', mask)
  if debug:
    print(dict(memory_bank))
  
  return sum(memory_bank.values())
```
