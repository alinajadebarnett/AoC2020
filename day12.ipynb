```
from itertools import *
from more_itertools import *
import numpy as np
import networkx as nx
import re
from sympy.ntheory import factorint
from collections import defaultdict
```

```
def main(input):
  pos = np.zeros(2)
  dir = np.array([1, 0])
  for line in input.split('\n'):
    if line[0] == 'N':
      pos += np.array([0, int(line[1:])])
    if line[0] == 'E':
      pos += np.array([int(line[1:]), 0])
    if line[0] == 'S':
      pos += np.array([0, -int(line[1:])])
    if line[0] == 'W':
      pos += np.array([-int(line[1:]), 0])
    if line[0] == 'L':
      angle = int(line[1:])*np.pi/180
      dir = np.array([[np.cos(angle), -np.sin(angle)], [np.sin(angle), np.cos(angle)]]).dot(dir)
    if line[0] == 'R':
      angle = -int(line[1:])*np.pi/180
      dir = np.array([[np.cos(angle), -np.sin(angle)], [np.sin(angle), np.cos(angle)]]).dot(dir)
    if line[0] == 'F':
      pos += int(line[1:]) * dir

  return np.abs(pos[0])+np.abs(pos[1])
```
