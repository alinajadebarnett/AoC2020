```
def step(program, machine_state, verbose=False):
  instr = program[machine_state['pc']]
  if instr[:3] == 'nop':
    machine_state['pc'] += 1
  elif instr[:3] == 'acc':
    machine_state['a'] += int(instr[4:])
    machine_state['pc'] += 1
  elif instr[:3] == 'jmp':
    machine_state['pc'] += int(instr[4:])
  if verbose:
    print(instr)
    print(machine_state)
```

```
def gen_repairs():
  for i, line in enumerate(program):
    if line[:3] == 'nop':
      yield program[:i] + [line.replace('nop', 'jmp')] + program[i+1:]
    if line[:3] == 'jmp':
      yield program[:i] + [line.replace('jmp', 'nop')] + program[i+1:]

print(next(gen_repairs())[:10])
print()
program[:10]
```
