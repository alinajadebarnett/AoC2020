```
def fun2(input, x, y):
  pos = 0
  trees = 0
  for lineno, line in enumerate(input.split('\n')):
    if lineno % y == 0:
      if line[pos] == '#':
        trees += 1
      pos += x
      if pos >= len(line):
        pos -= len(line)
  return trees
```
