```
from lark import Lark, UnexpectedCharacters
from string import ascii_letters
```

```
def get_language(input):
  lark_name_dict = {}
  for line in input.split('\n'):
    num = list(line.split(':'))[0]
    if int(num) == 0:
      target = 'start'
    elif ('a' in line):
      target = 'A'
    elif ('b' in line):
      target = 'B'
    else:
      target = num
      for number, letter in enumerate(ascii_letters):
        number = str(number)
        if len(number) == 1:
          target = target.replace(number, letter)
        else:
          break
    lark_name_dict[num] = target

  lark_replacement_list = list(lark_name_dict.items())
  lark_replacement_list.sort(key=lambda x: -len(x[0]))

  language = ''
  for line in input.split('\n'):
    for num, target in lark_replacement_list:
      line = line.replace(num, target)
    language += line + '\n'
  return language

def make_does_match(input):
  parser = Lark(get_language(input))
  def does_match(line):
    try:
      parser.parse(line)
    except UnexpectedCharacters as e:
      return False
    return True
  return does_match

def main(input, input2):
  does_match = make_does_match(input)
  s = 0
  for line in input2.split('\n'):
    if does_match(line):
      s += 1
  return s
```
