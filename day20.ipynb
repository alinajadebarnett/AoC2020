```
def fits(tile_center, tile_added, position):
  #print(type(tile_center), type(tile_added))
  if position == 'N':
    return np.array_equal(tile_center[0], tile_added[-1])
  if position == 'E':
    return np.array_equal(tile_center[:, -1], tile_added[:, 0])
  if position == 'S':
    return np.array_equal(tile_center[-1], tile_added[0])
  if position == 'W':
    return np.array_equal(tile_center[:, 0], tile_added[:, -1])
  else:
    raise ValueError('N, E, S, W')

def gen_orients(tile):
  for flip in (True, False):
    for rot_n in range(4):
      if flip:
        this_tile = tile.T
      else:
        this_tile = tile.copy()
      this_tile = np.rot90(this_tile, k=rot_n)
      yield this_tile, flip, rot_n

def is_definitely_corner(tile, tile_id, tile_dictionary):
  num_incompatible_sides = 0
  for position in ('N', 'E', 'S', 'W'):
    compatible = False
    for other_tile_id, other_tile in tile_dictionary.items():
      if not (tile_id == other_tile_id):
        for orient, _, _ in gen_orients(other_tile):
          if fits(tile, orient, position):
            compatible = True
            break
        if compatible:
          break
    if not compatible:
      num_incompatible_sides += 1
  return (num_incompatible_sides >= 2)

def main(input):
  s = 1
  tile_dict = {}
  for block in input.split('\n\n'):
    lines = list(block.split('\n'))
    block_number = int(lines[0].split(' ')[1][:-1])
    data = []
    for line in lines[1:]:
      data.append([])
      for c in line:
        if c == '#':
          data[-1].append(1)
        else:
          data[-1].append(0)
    data = np.array(data)
    tile_dict[block_number] = data

  corners = []
  for tile_id, tile in tile_dict.items():
    if is_definitely_corner(tile, tile_id, tile_dict):
      corners.append(tile)
      s *= tile_id

  return s
```

```
def fits(tile_center, tile_added, position):
  #print(type(tile_center), type(tile_added))
  if position == 'N':
    return np.array_equal(tile_center[0], tile_added[-1])
  if position == 'E':
    return np.array_equal(tile_center[:, -1], tile_added[:, 0])
  if position == 'S':
    return np.array_equal(tile_center[-1], tile_added[0])
  if position == 'W':
    return np.array_equal(tile_center[:, 0], tile_added[:, -1])
  else:
    raise ValueError('N, E, S, W')

def gen_orients(tile, tile_id):
  for flip in (True, False):
    for rot_n in range(4):
      if flip:
        this_tile = tile.T
      else:
        this_tile = tile.copy()
      this_tile = np.rot90(this_tile, k=rot_n)
      yield this_tile, tile_id, flip, rot_n

def find_incompatible_sides(tile, tile_id, tile_dictionary):
  num_incompatible_sides = 0
  incompatible_sides = []
  for position in ('N', 'E', 'S', 'W'):
    compatible = False
    for other_tile_id, other_tile in tile_dictionary.items():
      if not (tile_id == other_tile_id):
        for orient, _, _, _ in gen_orients(other_tile, other_tile_id):
          if fits(tile, orient, position):
            compatible = True
            break
        if compatible:
          break
    if not compatible:
      num_incompatible_sides += 1
      incompatible_sides.append(position)
  return num_incompatible_sides, incompatible_sides

def main(input, debug=True):
  tile_dict = {}
  for block in input.split('\n\n'):
    lines = list(block.split('\n'))
    block_number = int(lines[0].split(' ')[1][:-1])
    data = []
    for line in lines[1:]:
      data.append([])
      for c in line:
        if c == '#':
          data[-1].append(1)
        else:
          data[-1].append(0)
    data = np.array(data)
    tile_dict[block_number] = data

  corners = []
  edges = []
  interior = []
  for tile_id, tile in tile_dict.items():
    num_inc, which_inc = find_incompatible_sides(tile, tile_id, tile_dict)
    if num_inc == 2:
      corners.append((tile_id, which_inc))
    elif num_inc == 1:
      edges.append((tile_id, which_inc))
    elif num_inc == 0:
      interior.append((tile_id, which_inc))

  grid = [[None for i in range(12)] for j in range(12)]
  # this_tile, tile_id, flip, rot_n
  grid[0][0] = (tile_dict[2633], 2633, False, 0)

  added_tiles = set([2633])

  for row_num in range(len(grid)):
    for col_num in range(len(grid[0])):
      if debug:
        print("tile", row_num, col_num)
      if not (row_num==0 and col_num==0):
        if (row_num==0 or row_num==len(grid)-1) and (col_num==0 or col_num==len(grid)-1): #if at corner
          tile_possibilities = corners
        elif row_num==0 or row_num==len(grid)-1 or col_num==0 or col_num==len(grid)-1: #if at edge
          tile_possibilities = edges
        else: #if interior
          tile_possibilities = interior

        doubles_counter = 0
        for tile_id, _ in tile_possibilities:
          tile = tile_dict[tile_id]
          if not (tile_id in added_tiles):
            for this_tile, tile_id, flip, rot_n in gen_orients(tile, tile_id):
              fits_here = True
              if (col_num > 0 and not fits(tile_center=grid[row_num][col_num-1][0], tile_added=this_tile, position='E')):
                fits_here = False
              elif (row_num > 0 and not fits(tile_center=grid[row_num-1][col_num][0], tile_added=this_tile, position='S')):
                fits_here = False
              if fits_here:
                grid[row_num][col_num] = (this_tile, tile_id, flip, rot_n)
                if debug:
                  print(row_num, col_num, tile_id)
                added_tiles.add(tile_id)
                doubles_counter += 1
        if doubles_counter > 1:
          print("Warning: A place could fit multiple different tiles.")

  return grid
```
